# KNeighborsClassifier

```python
from sklearn.model_selection import train_test_split
x = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(weights='uniform',n_neighbors = 5)
model.fit(x_train,y_train)
pred = model.predict(x_test)
```

```python
from scipy.stats import zscore
KNN_s = model.score(x_test,y_test)*100
KNN_s 
```

# Support Vector Machine

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)
```

```python
pred_svc = svc.predict(x_test)
```

```python
SVC_s = svc.score(x_test,y_test)*100
SVC_s
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(x_train, y_train)
ran_pred=clf.predict(x_test)
RF_s = clf.score(x_test,y_test)*100
RF_s
```

# XGBoost

```python
import xgboost as xgb
clb = xgb.XGBClassifier(use_label_encoder=True,max_depth=30,gamma = 0.8)
clb.fit(x_train, y_train)
clb_pred = clb.predict(x_test)
XG_s =clb.score(x_test,y_test)*100
XG_s
```

```python
pred_XG = clb.predict(x_test)
```

```python
knn_pred = pd.DataFrame(pred)
svc_pred = pd.DataFrame(pred_svc)
ran_pred = pd.DataFrame(ran_pred)
XG_pred =  pd.DataFrame(pred_XG)
```

```python
result = pd.concat([knn_pred,svc_pred,ran_pred,XG_pred],axis=1)
```

# Performance Measures

```python
from sklearn.metrics import classification_report
```

```python
KNN_R = classification_report(y_test,pred)
print(KNN_R)
```

```python
SVC_R = classification_report(y_test,pred_svc)
print(SVC_R)
```

```python
RF_R = classification_report(y_test,ran_pred)
print(RF_R)
```

```python
XG_R = classification_report(y_test,pred_XG)
print(XG_R)
```

# Graphical Confusion Matrix

```python
from sklearn import metrics
conf_mat = metrics.confusion_matrix(y_test,pred,labels=[2,4])
data_confmat=pd.DataFrame(conf_mat,index=[i for i in [2,4]],columns=[i for i in ['predict B','predict M']] )
plt.figure(figsize=(3,2))
sns.heatmap(data_confmat,annot=True)
```
